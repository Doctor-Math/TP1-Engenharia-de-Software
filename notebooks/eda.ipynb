```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('../data/bankloans.csv')
```

```python
dados
```

```python
dados.isna().sum().sum() / dados.size * 100 #Nível de dados faltantes: 4% dos dados totais
```

```python
dados['default'].isna().sum() # 450 dados são NaN
```

```python
dados['default'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de Default')
plt.ylabel('')  # remove o label lateral desnecessário
plt.show()
```

```python
dados['age'].isna().sum()
```

```python
dados['age'].describe()
```

```python
sns.histplot(data=dados, x='age', bins=30, kde=True)
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
dados['ed'].isna().sum() 
```

```python
import matplotlib.ticker as ticker

sns.histplot(data=dados, x='ed', bins=5)
plt.title('Distribuição - Nível de Educação')
plt.xlabel('Nível de Educação')
plt.ylabel('Frequência')

plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.show()
```
